```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def LogisticMap(r, x0, N):
    n = np.arange(N)
    x_list = [x0]
    for i in range(N-1):
        xn = r * x_list[-1] * (1 - x_list[-1])
        x_list.append(xn)
 
    plt.figure(figsize=(9, 6))
    plt.scatter(n, x_list, facecolors='none', edgecolors='r')
    plt.plot(n, x_list)
    plt.ylabel('Xn', fontsize=16)
    plt.xlabel('N', fontsize=16)
    plt.title('Logistic Map for r= {} and x0= {}'.format(r, x0), fontsize=16)

    return None
```

```python
LogisticMap(3.57, 0.5, 50)
```

```python
LogisticMap(3.2, 0.5, 50)
```
